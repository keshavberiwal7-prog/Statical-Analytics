```python
age_groups = [(15, 25), (25, 35), (35, 45), (45, 55), (55, 65)]
frequencies = [350, 375, 120, 80, 90]

midpoints = [(start + end) / 2 for start, end in age_groups]

total = sum(frequencies)

mean = sum(f * m for f, m in zip(frequencies, midpoints)) / total

cum_freq = 0
for i, f in enumerate(frequencies):
    cum_freq += f
    if cum_freq >= total / 2:
        median_index = i
        break

L = age_groups[median_index][0] 
F = frequencies[median_index]  
CF = cum_freq - F      
w = age_groups[median_index][1] - age_groups[median_index][0] 

median = L + ((total / 2 - CF) / F) * w

mode_index = frequencies.index(max(frequencies))
L_mode = age_groups[mode_index][0]
f_mode = frequencies[mode_index]
f1 = frequencies[mode_index - 1] if mode_index > 0 else 0
f2 = frequencies[mode_index + 1] if mode_index < len(frequencies) - 1 else 0
w_mode = age_groups[mode_index][1] - age_groups[mode_index][0]

mode = L_mode + ((f_mode - f1) / ((2 * f_mode) - f1 - f2)) * w_mode

print(f"Mean = {mean:.2f}")
print(f"Median = {median:.2f}")
print(f"Mode = {mode:.2f}")
```
